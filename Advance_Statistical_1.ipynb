2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint
low = 1
high = 7
distribution = randint(low, high)
x = np.arange(low, high)
pmf = distribution.pmf(x)
plt.figure(figsize=(8, 5))
plt.stem(x, pmf, basefmt=" ",
         use_line_collection=True)
plt.title("Probability Mass Function of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
p = 0.7
dist = bernoulli(p)
x = [0, 1]
pmf = dist.pmf(x)
for xi, pi in zip(x, pmf):
  print(f"P(X = {xi}) = {pi}")
plt.figure(figsize=(6, 4))
plt.stem(x, pmf, basefmt=" ",
         use_line_collection=True)
plt.xticks(x)
plt.ylim(0, 1)
plt.xlabel('outcome')
plt.ylabel('Probability')
plt.title(f'Bernoulli Distribution (p = {p})')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

5.Create a Poisson distribution and visualize it using Python.

6.Write a Python program to calculate and plot the cumulative distributionfunction (CDF) of a discrete uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint
low = 1
high = 7
dist = randint(low, high)
x = np.arange(low - 1, high + 1)
cdf = dist.cdf(x)
plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='post',
         label='CDF', color='blue')
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(range(low - 1, high +1))
plt.yticks(np.linspace(0, 1, 7))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

9.Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
def calculate_and_plot_zscore(data):
  z_scores = zscore(data)
  print("First 10 Z-scores:",
        z_scores[:10])
  plt.figure(figsize=(8, 5))
  plt.hist(z_scores, bins=30, edgecolor='black', density=True, alpha=0.7)
  plt.title("Histogram of Z-score")
  plt.xlabel("Z-Score")
  plt.ylabel("Density")
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.show()
  return z_scores
```

12.Write a Python function to calculate and plot  the standard normal distribution (mean=0, std=1)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def plot_standard_normal():
  x = np.linspace(-4, 4, 1000)
  pdf = norm.pdf(x, loc=0, scale=1)
  plt.figure(figsize=(8, 5))
  plt.plot(x, pdf, color='blue', label='Standard Normal PDF')
  plt.title('Standard Normal Distribution (mean = 0, std = 1)')
  plt.xlabel('x')
  plt.ylabel('Probability Density')
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.legend()
  plt.show()
```

14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def z_scores(data_point, mean, std):
  z = (data_point - mean) / std
  print(f"Z-score for data point {data_point} is: {z:.4f}")
  return z
def plot_standard_normal_with_point(z):
  x = np.linspace(-4, 4, 1000)
  y = norm.pdf(x, loc=0, scale=1)
  plt.figure(figsize=(10, 5))
  plt.plot(x, y, label='Standard Normal Distribution', color='blue')
  plt.axvline(z, color='red', linestyle='--', label=f' z = {z:.2f}')
  plt.fill_between(x, y, 0, where=(x < z), color='red', alpha=0.3)
  plt.title('Z-Score Comparison to Standard Normal Distribution')
  plt.xlabel('Z')
  plt.ylabel('Probability Density')
  plt.legend()
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.show()
```

15.Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy import stats
sample_data = [52, 49, 51, 50, 53, 48, 47, 54, 50, 49]
pop_mean = 50
t_statistics, p_value = stats.ttest_1samp(sample-data, pop_mean)
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"T-Statistic: {t_statistics:.4f}")
print(f"P-Value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
  print("Reject the null hypothesis (significant difference).")
else:
  print("Fail to reject the null hypothesis (no significant difference).")
```

16.Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats
data = [52, 49, 51, 50, 53, 48, 47, 54, 50, 49]
mean = np.mean(data)
std_err = stats.sem(data)
confidence = 0.95
ci = stats.t.interval(confidence, df=len(data)-1, loc=mean, scale=std_err)
print(f"Sample Mean = {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval = ({ci[0]:.2f}, {ci[1]:.2f})")
```

17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=50)
sample_mean = np.mean(data)
std_err = stats.sem(data)
confidence = 0.95
ci = stats.t.interval(confidence, df=len(data)-1, loc=sample_mean, scale=std_err)
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci[0]:.2f}, {ci[1]:.2f})")
```

18.write a python script to calculate and visualize the probability denisty function (PDF) of a normal distribution.
